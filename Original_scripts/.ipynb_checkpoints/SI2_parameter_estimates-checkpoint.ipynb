```python
def blank_second(x, offset=0):
    out = []
    for i, xx in enumerate(x):
        if (i + offset) % 2 == 0:
            out.append(xx)
        else:
            out.append('')
    return out
```

```python
def figure_si2(estimates, 
               parameters = ['v', 's', 'gamma', 'tau'], 
               parameter_labels = dict(v='v',
                                       s=r'$\sigma$',
                                       gamma=r'$\gamma$',
                                       tau=r'$\tau$'),
               parameter_ticks = dict(v=np.arange(0, 0.00021, 0.00005),
                                      s=np.arange(0.004, 0.0151, 0.002),
                                      gamma=np.arange(-1, 1.1, 0.25),
                                      tau=np.arange(0, 6, 1)),
               parameter_ticklabels = dict(v=blank_second(np.arange(0, 0.00021, 0.00005)),
                                           s=blank_second(np.arange(0.004, 0.0151, 0.002), offset=1),
                                           gamma=blank_second(np.arange(-1, 1.1, 0.25)),
                                           tau=np.arange(0, 6, 1)),
               figsize = (12.5, 12.5), n_bins = 10, ci = 0.95, fontsize_axislabel = 12):

    n_parms = len(parameters)

    fig, axs = plt.subplots(n_parms, n_parms, figsize=figsize)

    for i, q in enumerate(parameters):
        for j, p in enumerate(parameters):

            if i < j:
                axs[i, j].axis('off')

            elif p == q:

                bins = np.linspace(parameter_ticks[p][0], parameter_ticks[p][-1], n_bins+1)
                axs[i, j].hist(estimates[p],
                               bins=bins,
                               normed=False, color='k', edgecolor='white')
                axs[i, j].set_xticks(parameter_ticks[p])
                axs[i, j].set_xticklabels(parameter_ticklabels[p])
                axs[i, j].set_xlabel(parameter_labels[p], fontsize=fontsize_axislabel)
                axs[i, j].set_ylabel('Frequency', fontsize=fontsize_axislabel)

            else:
                axs[i, j].scatter(estimates[p], estimates[q],
                                  s=16, facecolor='k', edgecolor='k', alpha=0.75)
                axs[i, j].set_xticks(parameter_ticks[p])
                axs[i, j].set_yticks(parameter_ticks[q])
                axs[i, j].set_xticklabels(parameter_ticklabels[p])
                axs[i, j].set_yticklabels(parameter_ticklabels[q])
                axs[i, j].set_xlim(parameter_ticks[p][0], parameter_ticks[p][-1])
                axs[i, j].set_ylim(parameter_ticks[q][0], parameter_ticks[q][-1])

                if i == (n_parms - 1):
                    axs[i, j].set_xlabel(parameter_labels[p], fontsize=fontsize_axislabel)
                if j == 0:
                    axs[i, j].set_ylabel(parameter_labels[q], fontsize=fontsize_axislabel)

                # Linear Model
                lm = run_linear_model(estimates[p], estimates[q], verbose=False)
                intercept, slope = lm.params
                table, data, columns = summary_table(lm, alpha=1.-ci)
                predicted, mean_ci_lower, mean_ci_upper = data[:, np.array([2, 4, 5])].T

                xs = np.linspace(*axs[i, j].get_xlim(), 100)
                axs[i, j].plot(xs, intercept + slope * xs,
                               color='r')
                sort_idx = np.argsort(estimates[p])
                axs[i, j].fill_between(estimates[p][sort_idx], mean_ci_lower[sort_idx], mean_ci_upper[sort_idx],
                                       color='red', alpha=0.1)

                # Annotation
                tval = lm.tvalues[-1]
                pval = lm.pvalues[-1]
                if pval < 0.01:
                    pstring = sci_notation(pval, decimal_digits=2)
                else:
                    pstring = str(np.round(pval, 2))
                r = np.sign(tval) * np.sqrt(lm.rsquared)
                axs[i, j].text(1.0, 1.0,
                               (r'$r = {:.2f}$, '.format(r)) +
                                r'$P =${}'.format(pstring),
                               verticalalignment='top',
                               horizontalalignment='right',
                               transform=axs[i, j].transAxes,
                               fontsize=10,
                               bbox=dict(facecolor='white', alpha=0.75))
    despine()
    fig.tight_layout()
    
    return fig, axs
```

```python
print('Close Figure to continue...')
figure_si2(estimates)
plt.savefig('results/figures/figure_si2_parameter_estimates.png', dpi=300, bbox_inches='tight')
plt.show()
```
